## 1. Which team scored the most points when playing at home?

```python
query = '''
SELECT * 
FROM Match;
'''
Match = pd.read_sql_query(query, connection)
Match
```

```python
Match.columns.to_list()
```

```python
Match.home_team_goal.value_counts()
```

```python
# On average the following teams scored more at home (summing up the scores might not be adequate since
# the number of games teams played can be different)
query = '''
SELECT AVG(Match.home_team_goal) AS home_team_goal_avg, Team.team_long_name
FROM Match
LEFT JOIN Team
ON Match.home_team_api_id=Team.team_api_id
GROUP BY Team.team_long_name
ORDER BY home_team_goal_avg DESC;
'''
scored_most_at_home = pd.read_sql_query(query, connection)
scored_most_at_home
```

```python
# As a sum the following teams scored more at home (keep in mind thast teams 
# might have played different number of games)
query = '''
SELECT SUM(Match.home_team_goal) AS home_team_goal_sum, Team.team_long_name
FROM Match
LEFT JOIN Team
ON Match.home_team_api_id=Team.team_api_id
GROUP BY Team.team_long_name
ORDER BY home_team_goal_sum DESC
LIMIT 10;
'''
scored_most_at_home_sum = pd.read_sql_query(query, connection)
scored_most_at_home_sum
```

## 2. Did this team also score the most points when playing away?

```python
# As a sum the following teams scored more at home (keep in mind thast teams 
# might have played different number of games)
query = '''
SELECT SUM(Match.away_team_goal) AS away_team_goal_sum, Team.team_long_name
FROM Match
LEFT JOIN Team
ON Match.home_team_api_id=Team.team_api_id
GROUP BY Team.team_long_name
ORDER BY away_team_goal_sum DESC
LIMIT 10;
'''
scored_most_at_away_sum = pd.read_sql_query(query, connection)
scored_most_at_away_sum
```

## 5. What was the median tie score? Use the value determined in the previous question for the number of tie games. Hint: PostgreSQL does not have a median function. Instead, think about the steps required to calculate a median and use the WITH command to store stepwise results as a table and then operate on these results

```python
query = '''
CREATE VIEW tie_table AS
SELECT home_team_goal AS ties
FROM Match
WHERE home_team_goal=away_team_goal
ORDER BY ties DESC;
'''
tie_table = pd.read_sql_query(query, connection)
```

```python
query = '''
select  max(ties) as [ties], quartile
from    (select ties, ntile(4) over (order by ties) as [quartile]
         from   tie_table) 
where quartile = 2
group by quartile
'''
tie_median = pd.read_sql_query(query, connection)
tie_median
```

```python
query = '''
SELECT x.ties 
FROM   (SELECT ties, 
               Count(1) OVER (partition BY 'A')        AS TotalRows, 
               Row_number() OVER (ORDER BY ties ASC) AS AmountOrder 
        FROM   tie_table ft) x 
WHERE  x.AmountOrder = Round(x.TotalRows / 2.0, 0)  
'''
tie_median = pd.read_sql_query(query, connection)
tie_median
```

## 6. What percentage of players prefer their left or right foot? Hint: Calculate either the right or left foot, whichever is easier based on how you setup the problem.

```python
query = '''
SELECT * 
FROM Player_Attributes;
'''
Player_Attributes = pd.read_sql_query(query, connection)
```

```python
Player_Attributes.columns
```

```python
query = '''
WITH left_count AS
(WITH left AS
(SELECT COUNT(preferred_foot) AS left
FROM Player_Attributes
WHERE preferred_foot='left'),
count AS
(SELECT COUNT(preferred_foot) AS count
FROM Player_Attributes)
SELECT left.left, count.count
FROM left, count)
SELECT CAST(left AS REAL) / count *100
FROM left_count;
'''
percent_preffered_left_foot = pd.read_sql_query(query, connection)
percent_preffered_left_foot
```
